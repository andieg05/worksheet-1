## Python Task 2: Parameter Sensitivity Analysis

Start task 2 once you have found parameter values that result in a graph that shows the characteristics of the oscillation (amplitude, decay rate, frequency, and phase shift).

Modify the code below to generate one additional figure that demonstrates the effect of varying one or more of the parameters in the model. Your figure should include an overlay of multiple plots. Make sure it is properly labeled with axis labels, a legend, and a descriptive title.

```python
import numpy as np
import matplotlib.pyplot as plt

# Use your parameters from Task 1 as a baseline
A = 2
k = 0.1
omega_base = 2 * np.pi # Renamed to avoid conflict when varying
phi = 0

# Set up time vector for the baseline
period_base = (2 * np.pi) / omega_base
t = np.linspace(0, 3 * period_base, 200)

# Create and format the multi-line plot with different parameter values
plt.figure(figsize=(10, 6))

# Baseline plot
y_base = A * np.exp(-k * t) * np.cos(omega_base * t + phi)
plt.plot(t, y_base, label='baseline')

# Add multiple plots with different parameters
# For example, to vary frequency (omega):
omega_values = [omega_base * 0.8, omega_base * 1.2]

for i, omega_var in enumerate(omega_values):
    y_varied = A * np.exp(-k * t) * np.cos(omega_var * t + phi)
    plt.plot(t, y_varied, linestyle='--', label='varied omega values')


# Add appropriate labels, title, legend, and grid
plt.title('Damped Oscillation: Sensitivity to Frequency') # Descriptive title
plt.xlabel('Time (s)') # Appropriate x-axis label
plt.ylabel('Amplitude') # Appropriate y-axis label
plt.grid(True)
plt.legend()

# Add text annotation to highlight important aspects of the parameter sensitivity
plt.text(0.5, 0.9, 'Higher $\omega$ results in faster oscillations', horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes, fontsize=10, bbox=dict(boxstyle="round,pad=0.3", fc="yellow", ec="b", lw=1, alpha=0.5))

plt.show()
```
